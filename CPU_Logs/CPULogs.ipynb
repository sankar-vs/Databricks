```
%sql
-- Checking Total Records in the table
SELECT user_name, count(*) AS records 
FROM CpuLogs  
GROUP BY user_name;
```

```
%sql
-- Finding users with highest number of average hours
/* Using from_unitxtime function we first calculate the Average by count(user_name) 
   As We know that the diff of every record is with difference of 5 min we multiply by 5
   We need the average time in Hours therefore we multiple by 60
   There are 6 days of data available, therefore we divide them by 6
   If there are 5 rows in data the difference between rows will result only 4 time (2-1, 3-2, 4-3, 5-4) we subtract 1 from the total count
   To get a accurate result we round the result to 2, and then covert it to time by using from_unittime function */
SELECT user_name, from_unixtime(round(((((count(user_name)-1)*5)*60)/6),2),'HH:mm') AS average_hours 
FROM CpuLogs
WHERE  keyboard != 0 OR mouse != 0  
GROUP BY user_name 
ORDER BY average_hours DESC;
```

```
%sql
-- Finding users with lowest number of average hours
/* Using from_unitxtime function we first calculate the Average by count(user_name) 
   As We know that the diff of every record is with difference of 5 min we multiply by 5
   We need the average time in Hours therefore we multiple by 60
   There are 6 days of data available, therefore we divide them by 6
   If there are 5 rows in data the difference between rows will result only 4 time (2-1, 3-2, 4-3, 5-4) we subtract 1 from the total count
   To get a accurate result we round the result to 2, and then covert it to time by using from_unittime function */
SELECT user_name, from_unixtime(round(((((count(user_name)-1)*5)*60)/6),2),'HH:mm') AS average_hours
FROM CpuLogs
WHERE  keyboard != 0 OR mouse != 0  
GROUP BY user_name 
ORDER BY average_hours ASC;
```

```
%sql
-- Finding users with highest numbers of idle hours
/* Using from_unitxtime function we first calculate the Average by count(user_name) 
   As We know that the diff of every record is with difference of 5 min we multiply by 5
   We need the average time in Hours therefore we multiple by 60
   There are 6 days of data available, therefore we divide them by 6
   If there are 5 rows in data the difference between rows will result only 4 time (2-1, 3-2, 4-3, 5-4) we subtract 1 from the total count
   To get a accurate result we round the result to 2, and then covert it to time by using from_unittime function */
SELECT user_name, count(*) AS numbers_of_times, from_unixtime(round(((((count(user_name)-1)*5)*60)/6),2),'HH:mm') AS idle_hours 
FROM CpuLogs
WHERE  keyboard == 0 AND mouse == 0  
GROUP BY user_name 
ORDER BY idle_hours DESC;
```

```
%sql
-- Finding users with highest numbers of times late comings
DROP TABLE IF EXISTS user_login_time;

CREATE TABLE user_login_time AS
SELECT to_date(DateTime) AS Date, Min(DateTime) AS min_time 
FROM CpuLogs 
GROUP BY Date 
ORDER BY Date;

DROP TABLE IF EXISTS user_active_time;

CREATE TABLE user_active_time AS
SELECT user_name, to_date(DateTime) AS Date, Min(DateTime) AS active_time 
FROM CpuLogs
WHERE keyboard != 0 OR mouse != 0 
GROUP BY user_name, Date 
ORDER BY Date;

SELECT td.user_name, from_unixtime(round((SUM(td.diff_time))/6,2),'HH:mm')AS avg_time, count(*) As no_of_times_late 
FROM(SELECT user_active_time.user_name, 
(unix_timestamp(user_active_time.active_time)- unix_timestamp(user_login_time.min_time)) AS diff_time
FROM user_active_time, user_login_time
WHERE user_active_time.Date = user_login_time.Date) AS td 
WHERE td.diff_time !=0 
GROUP BY td.user_name 
ORDER BY avg_time DESC;
```
